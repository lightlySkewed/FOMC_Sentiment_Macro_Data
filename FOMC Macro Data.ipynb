To get started, we need a list of official FOMC Meeting dates so that we can try to understand the world as they did on (or shortly before) the meeting date. This step is somewhat subjective

```python
url = 'https://fraser.stlouisfed.org/title/federal-open-market-committee-meeting-minutes-transcripts-documents-677?browse=1930s'
source = urllib.request.urlopen(url).read()
soup = bs.BeautifulSoup(source,'lxml')
# print(type(soup.find_all("script",type="application/ld+json")))
site_json=json.loads(soup.find("script",type="application/ld+json").text)
raw_meetings_list = []
for meeting in site_json['hasPart']:
    raw_meetings_list.append(meeting['name'])
```

```python
current_year = dt.datetime.today().year
years_list = list(range(2000, current_year, 1))
years_list_strings = [str(x) for x in years_list]
print(years_list_strings)

# keep items that start with Meeting, Telephone Conference, Videoconference
keep_regexes = ["^Meeting", "^Telephone Conference", "^Videoconference"]
keep_combined = "(" + ")|(".join(keep_regexes) + ")"

# drop items
drop_regexes = ["Cancelled", "Transcript", "Notation Vote"]
drop_combined = "(" + ")|(".join(drop_regexes) + ")"

# define items to keep list
keep_meeting_list = []

# add items to keep_meeting_list if they are of the correct criteria
for meeting in raw_meetings_list:
    if (
        re.search(keep_combined, meeting) and 
        any(year in meeting for year in years_list_strings) and not 
        re.search(drop_combined, meeting)
    ):
        # any throws True at the first match
        keep_meeting_list.append(meeting)

keep_meeting_list
```
